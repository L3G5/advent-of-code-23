```python
from collections import Counter
from dataclasses import dataclass

strength_single = {'A': 14, 'K': 13, 'Q': 12, 'J': 11, 'T': 10, '9': 9, '8': 8, '7': 7, '6': 6, '5': 5, '4': 4, '3': 3, '2': 2}
strength_hand = {'5': 6, '4': 5, 'FH': 4, '3': 3, '22': 2, '2': 1, '1': 0}

@dataclass(eq=False)
class Card:
    face: str
    def __gt__(self, other):
        return strength_single[self.face] > strength_single[other.face]
    def __lt__(self, other):
        return strength_single[self.face] < strength_single[other.face]
    def __eq__(self, other):
        return strength_single[self.face] == strength_single[other.face]

def counter_to_value(string):
    counts = Counter(string)
    if 5 in counts.values():
        return '5'
    if 4 in counts.values():
        return '4'
    if (3 in counts.values()) & (2 in counts.values()):
        return 'FH'
    if (3 in counts.values()):
        return '3'
    if Counter(counts.values())[2] == 2:
        return '22'
    if (2 in counts.values()):
        return '2'
    return '1'

class Hand:
    def __init__(self, string) -> None:
        self.hand = string
        self.cards = [Card(i) for i in string]
        self.value = strength_hand[counter_to_value(string)]
```

```python
def sort_hand(hands):
    return sorted(hands, key = lambda x: (x.value, x.cards[0], x.cards[1], x.cards[2], x.cards[3], x.cards[4]))
sorted_hands = sort_hand([Hand('ALKJQ'), Hand('AAAAA'), Hand('3AAKQ'), Hand('A22JQ')])
[i.hand for i in sorted_hands]
```

```python
def solution_1(text):
    hands_bids = [[Hand(i.split()[0]), int(i.split()[1])] for i in text]
    sorted_hand_bids = sorted(
    hands_bids,
    key = lambda x: (x[0].value, x[0].cards[0], x[0].cards[1], x[0].cards[2], x[0].cards[3], x[0].cards[4]))
    ans = 0
    for i, (hand, bid) in enumerate([(i[0].hand, i[1]) for i in sorted_hand_bids]):
        ans += (i+1)*bid
    # return ans
    print(ans)
    return [[i[0].hand, i[1]] for i in sorted_hand_bids]

print(solution_1(text_test))
print(solution_1(text))
```

```python
from collections import Counter
from dataclasses import dataclass

strength_single = {'A': 14, 'K': 13, 'Q': 12, 'J': 1, 'T': 10, '9': 9, '8': 8, '7': 7, '6': 6, '5': 5, '4': 4, '3': 3, '2': 2}
strength_hand = {'5': 6, '4': 5, 'FH': 4, '3': 3, '22': 2, '2': 1, '1': 0}
strength_hand_rev = {v: k for k, v in strength_hand.items()}

@dataclass(eq=False)
class Card:
    face: str
    def __gt__(self, other):
        return strength_single[self.face] > strength_single[other.face]
    def __lt__(self, other):
        return strength_single[self.face] < strength_single[other.face]
    def __eq__(self, other):
        return strength_single[self.face] == strength_single[other.face]

def counter_to_value_joker(string: str):
    counts = Counter(string)
    if 'J' not in counts.keys(): 
        return counter_to_value(string)
    if counts['J'] == 5:
        return '5'
    other_cards = [i for i in counts.keys() if i!= 'J']
    value = max([strength_hand[counter_to_value(string.replace('J', c))] for c in other_cards])
    return strength_hand_rev[value]
    
class HandJoker:
    def __init__(self, string) -> None:
        self.hand = string
        self.cards = [Card(i) for i in string]
        self.value = strength_hand[counter_to_value_joker(string)]
```
