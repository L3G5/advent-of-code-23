## part 1

```python
from collections import ChainMap
from itertools import cycle
```

```python
class Desert:
    def __init__(self, routes) -> None:
        self.routes = routes
        self.state = 'AAA'
    def move(self, d):
        if d == 'R':
            self.state = self.routes[self.state][1]
            return self.state
        if d == 'L':
            self.state = self.routes[self.state][0]
            return self.state
```

```python
def solution_1(text):
    path = text[0]
    location = Desert(dict(ChainMap(*[{i.split(' = ')[0]: i.split(' = ')[1].strip('()').split(', ')} for i in text[1].split('\n')])))

    ans = 0
    for d in cycle(path):
        location.move(d)
        ans += 1
        if location.state == 'ZZZ': break
    return ans

print(solution_1(text_test1))
print(solution_1(text_test2))
print(solution_1(text))
```

```python
class DesertGhost:
    def __init__(self, routes) -> None:
        self.routes = routes
        self.state = [st for st in routes.keys() if st[-1]=='A']
    def move(self, d):
        if d == 'R':
            self.state = [self.routes[st][1] for st in self.state]
            return self.state
        if d == 'L':
            self.state = [self.routes[st][0] for st in self.state]
            return self.state
```

```python
def solution_2(text):
    path = text[0]
    location = DesertGhost(dict(ChainMap(*[{i.split(' = ')[0]: i.split(' = ')[1].strip('()').split(', ')} for i in text[1].split('\n')])))
    print(location.state)
    ans = 0
    for d in cycle(path):
        location.move(d)
        ans += 1
        check = [st[-1]=='Z' for st in location.state]
        if any(check):
            print(location.state)
            print(check, ans, sep = '\t')
        if all(check): break
    return ans

# print(solution_2(text_test3))
# print(solution_2(text_test2))
print(solution_2(text))
```
